```python
df = df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'])
print(f"Columns after drop: {df.shape[1]}")
```

```python
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
print(df['Attrition_Flag'].value_counts())
```

```python
df['AgeGroup'] = pd.cut(df['Age'], 
                         bins=[18, 25, 35, 45, 60], 
                         labels=['18-25', '26-35', '36-45', '46-60'])
```

```python
df['SalaryBand'] = pd.cut(df['MonthlyIncome'],
                           bins=[0, 3000, 6000, 10000, 20000],
                           labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
df.to_csv('hr_clean.csv', index=False)
print("Clean dataset saved!")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('HR Analytics - Attrition Analysis', fontsize=16, fontweight='bold')

attrition_counts = df['Attrition'].value_counts()
axes[0,0].pie(attrition_counts, labels=attrition_counts.index, 
              autopct='%1.1f%%', colors=['#2ecc71','#e74c3c'])
axes[0,0].set_title('Overall Attrition Rate')

# Plot 2 — Department wise attrition
dept_attr = df.groupby('Department')['Attrition_Flag'].mean() * 100
dept_attr.plot(kind='bar', ax=axes[0,1], color='#3498db', edgecolor='black')
axes[0,1].set_title('Attrition Rate by Department (%)')
axes[0,1].set_xlabel('')
axes[0,1].tick_params(axis='x', rotation=0)

# Plot 3 — Overtime vs Attrition
ot_attr = df.groupby('OverTime')['Attrition_Flag'].mean() * 100
ot_attr.plot(kind='bar', ax=axes[0,2], color=['#2ecc71','#e74c3c'], edgecolor='black')
axes[0,2].set_title('Attrition Rate: Overtime vs No Overtime (%)')
axes[0,2].set_xlabel('')
axes[0,2].tick_params(axis='x', rotation=0)

# Plot 4 — Monthly Income vs Attrition
df.boxplot(column='MonthlyIncome', by='Attrition', ax=axes[1,0], 
           patch_artist=True)
axes[1,0].set_title('Monthly Income vs Attrition')
axes[1,0].set_xlabel('')

# Plot 5 — Job Satisfaction vs Attrition
js_attr = df.groupby('JobSatisfaction')['Attrition_Flag'].mean() * 100
js_attr.plot(kind='bar', ax=axes[1,1], color='#9b59b6', edgecolor='black')
axes[1,1].set_title('Attrition by Job Satisfaction (1=Low, 4=High)')
axes[1,1].set_xlabel('')
axes[1,1].tick_params(axis='x', rotation=0)

# Plot 6 — Age Group vs Attrition
age_attr = df.groupby('AgeGroup', observed=True)['Attrition_Flag'].mean() * 100
age_attr.plot(kind='bar', ax=axes[1,2], color='#e67e22', edgecolor='black')
axes[1,2].set_title('Attrition Rate by Age Group (%)')
axes[1,2].set_xlabel('')
axes[1,2].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('hr_eda_plots.png', dpi=150, bbox_inches='tight')
plt.show()
print("EDA plots saved!")
```

```python
print("=" * 50)
print("KEY INSIGHTS")
print("=" * 50)

# 1. Overall attrition rate
total_attr = df['Attrition_Flag'].mean() * 100
print(f"\n1. Overall Attrition Rate: {total_attr:.1f}%")

# 2. Overtime ka impact
ot_yes = df[df['OverTime']=='Yes']['Attrition_Flag'].mean() * 100
ot_no  = df[df['OverTime']=='No']['Attrition_Flag'].mean() * 100
print(f"\n2. Attrition with Overtime: {ot_yes:.1f}%")
print(f"   Attrition without Overtime: {ot_no:.1f}%")
print(f"   --> Overtime employees {ot_yes/ot_no:.1f}x more likely to leave")

# 3. Salary impact
avg_left   = df[df['Attrition']=='Yes']['MonthlyIncome'].mean()
avg_stayed = df[df['Attrition']=='No']['MonthlyIncome'].mean()
print(f"\n3. Avg Salary - Left: ${avg_left:.0f} | Stayed: ${avg_stayed:.0f}")

# 4. Department wise
dept_attr = df.groupby('Department')['Attrition_Flag'].mean() * 100
print(f"\n4. Highest Attrition Dept: {dept_attr.idxmax()} ({dept_attr.max():.1f}%)")

# 5. Distance from home
dist_left   = df[df['Attrition']=='Yes']['DistanceFromHome'].mean()
dist_stayed = df[df['Attrition']=='No']['DistanceFromHome'].mean()
print(f"\n5. Avg Distance - Left: {dist_left:.1f} km | Stayed: {dist_stayed:.1f} km")
```
